## 데이터셋의 특성
    1. SeriousDlqin2yrs:
        - 채무자가 지난 2년간 90일까지 기한 경과(past dues)를 경험했는지 여부를 나타낸다.
        - 경험이 있으면 Yes, 없으면 No
        - Target Label
    2. RevolvingUtilizationOfUnsecuredLines
        - 현재의 대출 채무(loan debt) 및 부동산(real estate)을 제외한 채무자의 신용카드 한도(credit card limits)를 나타낸다.
        - ex) 신용한도 1,000달러, 은행계좌 1,000달러, 신용카드 잔고 1,000달러중 500달러 일 때, 최대 잔고는 2,000달러 이고 현재 
              잔고는 1,00달러이므로 특성 값은 1,500 / 2,000 = 0.75가 된다
    3. Age
        - 채무자의 나이
    4. NumberOfTime30-59DaysPastDueNotWorse
        - 만기일로부터 30일 ~ 59일내에 지불한 횟수
    5. DebtRatio
        - 부채 비율
        - ex) 월간 부채 200달러, 그 외 지출 500달러, 월간 소득 1,000달러 일 때, 부채비율은 700 / 1,000 = 0.7
    6. MonthlyIncome
        - 채무자의 월간 소득
    7. NumberOfOpenCreditLinesAndLoans
        - 자유 상환 대출(Open Loans) 개수 및 채무자가 유지하는 신용카드 개수
    8. NumberOfTimes90DaysLate
        - EMI를 지급한 날로부터 90일 후에 미지급금을 지불한 횟수
    9. NumberRealEstateLoansOrLines
        - 부동산 담보 대출 개수 또는 주택 담보 대출 개수
    10. NumberOfTime60-89DaysPastDueNotWorse
        - 만기일로부터 60일 ~ 89일 내에 지불한 횟수
    11. NumberOfDependents
        - 채무자의 부양 가족(dependents)수
        - 부양 가족 수에서 채무자는 제외

## 이상점(outliers) 검출 기법
    1. 백분위수(percentile)
    2. 중위 절대 편차(median absolut deviation)
    3. 표준편차(standard deviation)
    4. 다수결 투표(majority-vote)
    5. 이상점의 시각화(visualization)

```python
# 백분위수 기반의 이상점 검출
# 백분위수 범위가 2.5 ~ 97.5 사이인 모든 데이터 점을 사용한다고 가정
# 95라는 분계점(threshold) 정해 백분위수 범위를 도출
def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2.0
    (minval, maxval) = np.percentile(data, [diff, 100 - diff])
    return ((data < minval) | (data > maxval))
```

```python
# 중위 절대 편차(수정 Z 점수) 기반 이상점 검출
# 1. 중위수를 찾는다
# 2. 각 값에 중위수를 빼고 절대값을 찾는다
# 3. 2번째에서 얻은 값들에서 중위값을 찾는다 - MAD값
# 4. (0.6745 * (x - |x|)) / MAD

def mad_based_outlier(points, threshold=3.5):
    if len(points.shape) == 1:
        points = points[:None]
    median_y = np.median(points)
    median_absolute_deviation_y = np.median([np.abs(y - median_y) for y in points])
    modified_z_scores = [(0.6745 * (y - median_y)) / (median_absolute_deviation_y+1e-7) for y in points]
    
    return np.abs(modified_z_scores) > threshold
```

```python
# 표준편차 기반 이상점 검출
def std_div(data, threshold=3):
    std = data.std()
    mean = data.mean()
    isOutlier = []
    for val in data:
        if val/std > threshold:
            isOutlier.append(True)
        else:
            isOutlier.append(False)
    return isOutlier
```

```python
# 다수결 투표 기반 이상점 검출
# 다른 이상점 검출방식을 사용하여 값을 구하고 구한 값들의 다수결로 선택
def outlierVote(data):
    x = percentile_based_outlier(data)
    y = mad_based_outlier(data)
    z = std_div(data)
    temp = list(zip(data.index, x, y, z))
    final = []
    for i in range(len(temp)):
        if temp[i].count(False) >= 2:
            final.append(False)
        else:
            final.append(True)
    return final
```

```python
# 이상점 시각화
def plotOutlier(x):
    fig, axes = plt.subplots(nrows=4)
    for ax, func in zip(axes, [percentile_based_outlier, mad_based_outlier, std_div, outlierVote]):
        sns.kdeplot(x, ax=ax)
        outliers = x[func(x)]
        ax.plot(outliers, np.zeros_like(outliers), 'ro', clip_on=False)

    kwargs = dict(y=0.95, x=0.05, ha='left', va='top', size=20)
    axes[0].set_title('Percentile-based Outliers', **kwargs)
    axes[1].set_title('MAD-based Outliers', **kwargs)
    axes[2].set_title('STD-based Outliers', **kwargs)
    axes[3].set_title('Majority vote based Outliers', **kwargs)
    fig.suptitle('Comparing Outlier Tests with n={}'.format(len(x)), size=20)
    fig = plt.gcf()
    fig.set_size_inches(15,10)
```

```python
# 각 방법에서 파생된 이상점의 수를 세는 것만으로 비교
def outlierRatio(data):
    functions = [percentile_based_outlier, mad_based_outlier, std_div, outlierVote]
    outlierDict = {}
    for func in functions:
        funcResult = func(data)
        count = 0
        for val in funcResult:
            if val == True:
                count += 1
        outlierDict[str(func)[10:].split()[0]] = [count, '{:.2f}%'.format((float(count)/len(data))*100)]
    return outlierDict
outlierRatio(training_data.debtratio)
```

```python
# 이상점의 최대 수는 MAD 기반 방법으로 검출되었으므로 이 방법을 고려할 생각
# 이상값 대체를 위한 최소 상계 값(minimum upper bound)찾기
# Minimum Upper Bound: 이상값에서 파생된 최솟값
minUpperBound = min([val for (val, out) in zip(training_data.debtratio, mad_based_outlier(training_data.debtratio)) if out == True])
```

```python
newDebtRatio = []
for val in training_data.debtratio:
    if val > minUpperBound:
        newDebtRatio.append(minUpperBound)
    else:
        newDebtRatio.append(val)
training_data.debtratio = newDebtRatio
```

```python
def replaceOutlier(data, method = outlierVote, replace='median'):
    vote = outlierVote(data)
    x = pd.DataFrame(zip(data, vote), columns=['debt', 'outlier'])
    if replace == 'median':
        replace = x.debt.median()
        print('median')
    elif replace == 'minUpper':
        replace = min([val for (val, vote) in zip(data, vote) if vote == True])
        if replace < data.mean():
            return 'There are outliers lower than the sample mean'
        print('minUpper')
    debtNew = []
    print(x.shape[0], replace)
    for i in range(x.shape[0]):
        if x.iloc[i][1] == True:
            debtNew.append(replace)
        else:
            debtNew.append(x.iloc[i][0])
            
    return debtNew
```

## 테스트 계량
    1. 훈련된 모델의 평균 정확도(mean accuracy)
        - 사이킷런이 제공하는 score()함수 사용
        - 잔차 표준오차(residual standard error)를 반환
        - 잔차 표준오차는 평균 제곱 오차(mean square error)의 양의 제곱근
        - 평균 제곱오차는 1 - (sum((target_label - predicted_label)**2) / sum((target_label - target_mean_label)**2))
    2. ROC-AUC 점수
        - 분류기의 정확도 확인에 사용
        - ROC와 AUC는 서로 다른 용어
        1) ROC(receiver operating characteristic)
            - 수신기 검사 특성
            - 일종의 2차원 곡선 모양
            - x 축은 거짓 양성 비율(false positive rate, FPR)
            - y 축은 참 양성 비율(true positive rate, TPR)
            - TPR은 민감도(sensitivity)라고도 불리며, 참 양성 / 양성 표본 개수 = TP/P로 구한다
            - FPR은 특이도(specificity,SPC)라고도 불리며, 거짓 양성 / 음성 표본 개수 = FP/N = 1-SPC로 구한다
        2) AUC(area under the curve)
            - ROC 곡선을 정량화하기 위해 사용
            - ROC 곡선이 얼마나 면적을 차지했는지 판별
            - 점수가 0.6 ~ 0.9이면 좋은 분류기라 간주

## 기존 접근법의 문제점
    1. 과전합 문제를 확인하기위해 교차 검증(cross-validation)을 사용하지 않았다.
    2. 하이퍼파라미터 조율(hyperparameter tuning)에 초점을 맞추지 않았다

## 접근법 최적화
    1. 교차 검증(cross validation)
        - 훈련-테스트 분할 접근법은 실제로는 훈련을 위해 전체 훈련 데이터세을 사용하지 않아
          훈련 데이터셋에 있는 모든 상황을 파악하지 못할 수 있다는 문제점이 있다.
        - 위 문제점을 해결한 방법이 K겹 교차 검증이다.
        1) K겹 교차 검증(K-fold CV)
            - 훈련 데이터셋을 k개의 하위 데이터셋으로 나눈다.
            - k - 1의 수만큼 반복과정을 실행한다.
            - 처음에 테스트를 위해 한 개의 부분집합을 취하고 훈련을 위해 나머지 네 개의 부분집합을 취한다.
            - 다음 테스트를 위해 두 개의 부분집합을 취하고 훈련을 위해 나머지 세 개 부분집합은 훈련용으로 사용한다.
            - ... 반복
            - 테스트를 위해 남아 있는 부분집합이 없는 k - 1회 반복 후 중단한다.
            - k겹 교차 검증은 모든 데이터 점을 훈련에 사용한다는 점에 이점을 누린다.
            - 모든 반복과정이 끝나고 정확도 점수를 얻는다.
            - 모든 반복과정이 완료되면 정확도 점수의 평균값을 취해 정확도 점수에서 표준편차값을 도출
            - 평균 및 표준편차 점수는 과적합 여부로 인한 영향을 식별하는데 도움이 된다.
            - 계산 비용이 크다는 단점이 있다.
            
    2. 하이퍼파라미터 조율(hyperparameter tuning)
        - 모델 훈련 시 모델이 학습할 수 없는 파라미터
        - 사용자가 직접 설정하여 모델의 성능을 조율(tuning)할 수 있는 파라미터
        1) 격자탐색(grid search) 파라미터 조율
            - 격자(grid)라고 부르는 리스트에 파라미터 값들을 지정
            - 격자에 지정된 각 값은 파라미터 조정 중에 고려해야 할 대상
            - 지정된 격자 값을 기반으로 구축하고 평가
            - 모든 파라미터 조합을 철저히 고려하고 최적 파라미터를 최종적으로 생성
            - 많은 시간이 필요하다는 단점이 있다.
        2) 임의탐색(random search) 파라미터 조율
            - 격자탐색과 동일한 방식
            - 차이점은 격자의 부분집합을 선택해 파라미터를 임의로 선택한다는 점

## 교차 검증 기반 접근법

```python
from sklearn.model_selection import cross_val_score
def cvDictGen(functions, scr, X_train=X, y_train=y, cv=5, verbose=1):
    cvDict = {}
    for func in functions:
        cvScore = cross_val_score(func, X_train, y_train, cv=cv, verbose=verbose, scoring=scr)
        cvDict[str(func).split('(')[0]] = [cvScore.mean(), cvScore.std()]
        
    return cvDict

def cvDictNormalize(cvDict):
    cvDictNormalized = {}
    for key in cvDict.keys():
        for i in cvDict[key]:
            cvDictNormalized[key] = ['{:0.2f}'.format((cvDict[key][0]/cvDict[cvDict.keys()[0]][0])),
                                     '{:0.2f}'.format((cvDict[key][1]/cvDict[cvDict.keys()[0]][1]))]
    return cvDictNormalized
```

```python
cvD = cvDictGen(functions=[knMod, glmMod, adaMod, gbMod, rfMod], scr='roc_auc')
cvD
```
